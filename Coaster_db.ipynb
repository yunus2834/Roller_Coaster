# 5  - Ask a Questions about the Data

```python
df.query('Location != "Other"') \
    .groupby("Location")["Speed_Mph"] \
    .agg(["mean","count"]) \
    .query("count >= 10") \
    .sort_values("count") \
     .plot(kind="barh")   #["mean"] # 10 dan fazla hız trenine sahip şehirlerin ortalama hızları
```

```python
df.query('Location != "Other"') \
    .groupby("Location")["Speed_Mph"].agg(["count"]).query("count >= 10") \
.sort_values("count",ascending=False).head(10)
```

```python
df.query('Location != "Other"') \
.groupby("Location")["Speed_Mph"].agg(["count"]).nlargest(10,"count")
```

```python
df.query('Location != "Other"') \
    .groupby("Location")["Speed_Mph"].agg(["count"]).query("count >= 10").sort_values("count").head(10).plot(kind="barh")
```

```python
df.query('Location != "Other"') \
.groupby("Location")["Speed_Mph"].agg(["count"]).nlargest(10,"count").plot(kind="barh")
```

```python
df.query('Location != "Other"') \
    .groupby("Location")["Speed_Mph"] \
    .agg(["mean","count"]) \
    .query("count >= 10") \
    .sort_values("mean")["mean"] \
     .plot(kind="barh",figsize=(12,5),title="Average Coast Speed by Location")  
#["mean"] # 10 dan fazla hız trenine sahip şehirlerin ortalama hızları

plt.show()
```
